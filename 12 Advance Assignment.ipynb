Q1. Does assigning a value to a string's indexed character violate Python's string immutability?

No, assigning a value to a string's indexed character does not violate Python's string immutability. Python strings are immutable, which means that once a string is created, its contents cannot be changed. However, you can use string concatenation or slicing to create a new string with the desired modifications.

Q2. Does using the += operator to concatenate strings violate Python's string immutability? Why or why not?

No, using the += operator to concatenate strings does not violate Python's string immutability. Although strings are immutable in Python, the += operator does not modify the original string. Instead, it creates a new string by concatenating the original string with the assigned value and reassigns it to the same variable. This operation does not change the original string itself, but rather creates a new string object. Therefore, the += operator is allowed for string concatenation without violating string immutability.

Q8. What is the benefit of first checking the target string with in or not in before using the index method to find a substring?

Checking the target string with the in or not in operator before using the index() method to find a substring offers a benefit of avoiding potential errors or exceptions. The in and not in operators allow you to check if a substring exists within a larger string. By performing this check first, you can prevent a ValueError that would occur if the index() method is used directly on a substring that does not exist in the target string

Q9. Which operators and built-in string methods produce simple Boolean (true/false) results?

Operators:

Equality operator (==): It compares two strings and returns True if they are equal, False otherwise.
Example: "hello" == "hello" returns True.

Inequality operator (!=): It compares two strings and returns True if they are not equal, False otherwise.
Example: "hello" != "world" returns True.

Built-in string methods:

startswith(): It checks if a string starts with a specified prefix and returns True or False accordingly.
Example: "Hello, world!".startswith("Hello") returns True.

endswith(): It checks if a string ends with a specified suffix and returns True or False accordingly.
Example: "Hello, world!".endswith("world!") returns True.

isalpha(): It checks if a string contains only alphabetic characters (letters) and returns True or False accordingly.
Example: "Hello".isalpha() returns True, while "Hello123".isalpha() returns False.

isdigit(): It checks if a string contains only numeric characters and returns True or False accordingly.
Example: "123".isdigit() returns True, while "Hello123".isdigit() returns False.

islower(): It checks if all alphabetic characters in a string are lowercase and returns True or False accordingly.
Example: "hello".islower() returns True, while "Hello".islower() returns False.

isupper(): It checks if all alphabetic characters in a string are uppercase and returns True or False accordingly.
Example: "HELLO".isupper() returns True, while "Hello".isupper() returns False.
